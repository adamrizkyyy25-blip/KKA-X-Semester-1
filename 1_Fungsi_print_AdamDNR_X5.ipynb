#5. Parameter Fungsi Print

```python
nama_siswa_1 = "Jielo"
nama_siswa_2 = "Aida"
nama_siswa_3 = "Dani"
nama_siswa_4 = "Nurul"

print(nama_siswa_4, end="/")
print(nama_siswa_2, end="/")
print(nama_siswa_1, end="/")
print(nama_siswa_3)
```
